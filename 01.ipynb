```python
import numpy as np

f = open("01_input.txt","r")
input = f.readlines()
input = [int(x.split('\n')[0]) for x in input]
```

```python
status = np.zeros(len(input))
print(len(input))

for idx, meter in enumerate(input):
    if idx == 0:
        status[0] = 0
    elif meter > input[idx - 1]:
        status[idx] = 1
    else:
        status[idx] = 0

print(np.sum(status))
```

```python
status = np.zeros(len(input))
old_sum = 0

for idx, meter in enumerate(input):

    if idx < len(input) -2:
        sum = input[idx] + input[idx+1] + input[idx+2]
    if idx == 0:
        status[0] = 0
    elif sum > old_sum:
        status[idx] = 1
    else:
        status[idx] = 0
    old_sum = sum
print(np.sum(status))
```

```python
f = open("02_input.txt","r")
input = f.readlines()

directions = []
for line in input:
    dir = line.split(' ')
    directions.append((dir[0],int(dir[1].split('\n')[0])))
```

```python
submarine_coord = [0,0]

for dir, coord in directions:
    if dir == 'forward':
        submarine_coord[0] += coord
    elif dir == "down":
        submarine_coord[1] += coord
    elif dir == "up":
        submarine_coord[1] -= coord
print(submarine_coord[0]*submarine_coord[1])

```

```python
submarine_coord = [0,0,0]
#horizontal, depth, aim

for dir, coord in directions:
    if dir == 'forward':
        #increase horizontal & depth = aim * x
        submarine_coord[0] += coord
        submarine_coord[1] += submarine_coord[2] * coord
    elif dir == "down":
        submarine_coord[2] += coord
    elif dir == "up":
        submarine_coord[2] -= coord
print(submarine_coord[0]*submarine_coord[1])

```

```python
import numpy as np

f = open("03_input.txt","r")
input = f.readlines()
input = [x.split('\n')[0] for x in input]

m = np.zeros((12,len(input)))
for idx, number in enumerate(input):
    for i, digit in enumerate(number):
        m[i][idx] = int(digit)

m_sum = np.sum(m, axis = 1)
print(m_sum)

gamma_rate = [1 if x >= len(input)/2 else 0 for x in m_sum]
print(gamma_rate)
epsilon_rate = [0 if x > len(input)/2 else 1 for x in m_sum]
print(epsilon_rate)

def toDec(arr):
    dec = 0
    for idx, digit in enumerate(arr):
        if digit:
            dec += 2**(len(arr) - 1 - idx)
    return dec

print(toDec(gamma_rate))
print(toDec(epsilon_rate))

print(toDec(epsilon_rate) * toDec(gamma_rate))
```

```python
def filter(matrix, curr_val, curr_val_idx):
    filtered_matrix = np.zeros((0,12))

    for idx, row in enumerate(matrix):
        if row[curr_val_idx] != curr_val:
            filtered_matrix = np.vstack((filtered_matrix, row))
    return filtered_matrix
    
def mostCommon(matrix, idx):
    matrix = matrix.T
    sumAtIdx = np.sum(matrix, axis = 1)[idx]
     
    retVal = 1 if sumAtIdx >= np.shape(matrix)[1]/2 else 0
    return retVal 

def leastCommon(matrix, idx):
    matrix = matrix.T

    sumAtIdx = np.sum(matrix, axis = 1)[idx]
     
    retVal = 0 if sumAtIdx >= np.shape(matrix)[1]/2 else 1
    return retVal 

def oxygen(matrix):
    matrix = matrix.T
    
    for i in range(0,12):
        if np.shape(matrix) != (1,12):
            mostCommonBit = mostCommon(matrix, i)
            filteredMatrix = filter(matrix, mostCommonBit, i)
            matrix = filteredMatrix
    return matrix

def co2(matrix):
    matrix = matrix.T
    
    for i in range(0,12):
        if np.shape(matrix) != (1,12):
            leastCommonBit = leastCommon(matrix, i)
            filteredMatrix = filter(matrix, leastCommonBit, i)
            matrix = filteredMatrix
    return matrix

o = oxygen(m)
print(toDec(o[0]))
c = co2(m)
print(toDec(c[0]))
print(toDec(o[0]) * toDec(c[0]))
```
